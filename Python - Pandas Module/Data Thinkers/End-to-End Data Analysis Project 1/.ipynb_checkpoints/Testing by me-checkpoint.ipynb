```python
df1.drop(['Row ID', 'Order ID'], axis=1, inplace=True)
```

```python
cat_profit=df1.groupby("Category")["Profit"].sum()
cat_profit.plot(kind="bar")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
cat_profit=df1.groupby("Region")["Sales"].sum()
cat_profit.plot(kind="bar")
plt.title("Sale by Region")
plt.xlabel("Region")
plt.ylabel("Total Sale")
plt.show()
```

```python
df1["Order_Month"] = pd.DatetimeIndex(df1["Order Date"]).month

month_sales = df1.groupby("Order_Month")["Sales"].sum()
month_sales.plot(kind="line")
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sale")
plt.show()
```

```python
orders_by_shipping_mode = df1.groupby("Ship Mode").size()

returned_order_by_shipping_mode = df1[df1["Profit"] < 0].groupby("Ship Mode").size()

percentage_returned_order_by_shipping_mode = (returned_order_by_shipping_mode/orders_by_shipping_mode)*100

percentage_returned_order_by_shipping_mode.plot(kind="bar")
plt.title("Returned Percentage by Shipping Mode")
plt.xlabel("Shipping mode")
plt.ylabel("Return Percentage")
plt.show()
```

```python
df1["Order Day"] = pd.DatetimeIndex(df1["Order Date"]).day_name()

day_profit = df1.groupby("Order Day")["Profit"].sum()

day_profit.plot(kind="bar")
plt.title("Total Profit by the day of the week")
plt.xlabel("Day of the week")
plt.ylabel("Total Profit")
plt.show()
```
